```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Knn_Customers.csv")
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.25,random_state=0)
```

```python
#Scaling important even for 0/1 values
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# k tuning
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
accuracy_list = []
k_values = range(1, 8)
```

```python
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracy_list.append(accuracy_score(y_test, y_pred))
```

```python
# plot
plt.plot(k_values, accuracy_list, marker = 'o')
plt.xlabel("k value")
plt.ylabel("Accuracy")
plt.title("KNN - Choosing Best K")
plt.show()
```

```python
best_k = k_values[np.argmax(accuracy_list)]
print("Best K:",best_k)
```

```python
knn_final =  KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(x_train, y_train)

y_pred=knn_final.predict(x_test)

df= pd.DataFrame({'Actual': y_test,'Predicted': y_pred})
print(df)
```

```python
import joblib

joblib.dump(knn_final, "knn_customer_model.pkl")
joblib.dump(sc , "scaler.pkl")

print("KNN model and scaler saved successfully!")
```
